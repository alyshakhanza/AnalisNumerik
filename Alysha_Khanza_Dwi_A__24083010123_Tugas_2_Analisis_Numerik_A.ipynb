```python
# NAMA: Alysha Khanza Dwi Avianti
# NPM : 24083010123

import numpy as np
import matplotlib.pyplot as plt
# interval
x = np.linspace(0, 10, 500)
# deret fungsi sinusoidal
f = np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)
```

```python
# Hitung integral dan diferensial dari fungsi â€˜fâ€™ untuk interval x
# Turunan analitik
f_prime = np.cos(x) + np.cos(2*x) + np.cos(3*x) + np.cos(4*x) + np.cos(5*x)

# Integral analitik (antiturunan, tanpa C)
F = -(np.cos(x)) - (1/4)*np.cos(2*x) - (1/9)*np.cos(3*x) - (1/16)*np.cos(4*x) - (1/25)*np.cos(5*x)

# Integral tentu pada interval [0,10]
I_total = F[-1] - F[0]
print("Integral tentu âˆ«0^10 f(x) dx =", I_total)
```

dilakukan perhitungan turunan analitik
ð‘“â€²(x) dan integral analitik F(x) dari fungsi ð‘“(ð‘¥). Kemudian dihitung integral tentu pada interval [0,10], hasilnya â‰ˆ 2.1866

```python
# Hitung secara terpisah bagian integral di atas dan di bawah y=0
# Pisahkan f(x) positif & negatif
f_pos = np.where(f > 0, f, 0)
f_neg = np.where(f < 0, f, 0)

# Hitung luas area dengan metode trapezoid
area_above = np.trapezoid(f_pos, x)          # area di atas y=0
area_below_mag = -np.trapezoid(f_neg, x)     # area di bawah y=0 (positif)

print("Luas area di atas y=0 =", area_above)
print("Luas area di bawah y=0 =", area_below_mag)
print("Integral total (cek) =", area_above - area_below_mag)
```

Memisahkan bagian fungsi yang bernilai positif dan negatif, kemudian menghitung luas area masing-masing menggunakan metode trapezoid. Hasilnya:

Luas area positif â‰ˆ 4.6044

Luas area negatif â‰ˆ 2.4178

Selisih keduanya â‰ˆ 2.1866, sama dengan hasil integral tentu (validasi benar).

```python
# Plot semua fungsi: f, fâ€™, F (atas dan bawah dengan area warna berbeda)

plt.figure(figsize=(12, 8))

# Plot f(x) dengan area di atas & bawah
plt.subplot(3, 1, 1)
plt.plot(x, f, color="blue", label="f(x)")
plt.fill_between(x, f, 0, where=(f > 0), color="green", alpha=0.3, label="f(x) > 0")
plt.fill_between(x, f, 0, where=(f < 0), color="red", alpha=0.3, label="f(x) < 0")
plt.axhline(0, color="black", linewidth=1, linestyle="--")
plt.title("Fungsi f(x) dengan Area Positif & Negatif")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.legend(); plt.grid(True)

# Plot f'(x)
plt.subplot(3, 1, 2)
plt.plot(x, f_prime, color="orange", label="fâ€²(x)")
plt.axhline(0, color="black", linewidth=1, linestyle="--")
plt.title("Turunan f(x)")
plt.xlabel("x"); plt.ylabel("fâ€²(x)")
plt.legend(); plt.grid(True)

# Plot F(x)
plt.subplot(3, 1, 3)
plt.plot(x, F, color="purple", label="F(x) = âˆ«f(x) dx + C")
plt.axhline(0, color="black", linewidth=1, linestyle="--")
plt.title("Antiturunan F(x)")
plt.xlabel("x"); plt.ylabel("F(x)")
plt.legend(); plt.grid(True)

plt.tight_layout()
plt.show()
```

Subplot 1: f(x) dengan area hijau (f(x)>0) dan merah (f(x)<0).

Subplot 2: Turunan fâ€²(x).

Subplot 3: Antiturunan F(x).

Grafik menunjukkan bahwa turunan fâ€²(x) lebih sering berosilasi karena melibatkan fungsi cosinus, sedangkan integral F(x) cenderung lebih halus.
