```python
import numpy as np
import matplotlib.pyplot as plt

#Data
n = 200
NPM = 123
np.random.seed(NPM)

x = np.random.uniform(0, 4*np.pi, n)
y_true = 4 * np.sin(2*x + 0.2)  # Fungsi asli
noise = np.random.normal(0, 0.4, n)
y_noisy = y_true + noise

#Urutkan data agar rapi
idx = np.argsort(x)
x_sorted = x[idx]
y_sorted = y_noisy[idx]

# Grid untuk plotting
x_new = np.linspace(0, 4*np.pi, 500)
y_true_curve = 4 * np.sin(2*x_new + 0.2)

# Interpolasi Polinomial
polys = {}
for deg in [1, 2, 3, 5]:  # Linear, Kuadrat, Kubik, dan Orde 5
    coeffs = np.polyfit(x_sorted, y_sorted, deg)
    p = np.poly1d(coeffs)
    polys[deg] = p

# --- Plot ---
plt.figure(figsize=(10,6))
plt.scatter(x, y_noisy, color="red", alpha=0.5, label="Noisy data")
plt.plot(x_new, y_true_curve, "b-", linewidth=2, label="True function")

colors = {1:"green", 2:"purple", 3:"orange", 5:"black"}
styles = {1:"-", 2:"--", 3:"-.", 5:":"}

for deg, p in polys.items():
    plt.plot(x_new, p(x_new), styles[deg], color=colors[deg],
             linewidth=2, label=f"Polynomial degree {deg}")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolasi Polinomial (Linear, Kuadrat, Kubik, Orde 5)")
plt.legend()
plt.grid(True)
plt.show()

#  Cetak persamaan polinom
for deg, p in polys.items():
    print(f"Polinom orde {deg}:")
    print(p)
    print()
```

### Analisis Interpolasi Linear, Kuadrat, Kubik, dan Orde 5
- Orde 1 (Linear): hanya berupa garis lurus, tidak mengikuti bentuk sinusoidal.
- Orde 2 (Kuadrat): sedikit lebih baik, tetapi tetap jauh dari fungsi asli.
- Orde 3 (Kubik): mulai fleksibel, namun masih tidak mampu menggambarkan pola sinus.
- Orde 5: jauh lebih mendekati fungsi asli meskipun masih ada pengaruh noise.


```python

import numpy as np
import matplotlib.pyplot as plt

# Data
n = 200
NPM = 123
np.random.seed(NPM)

x = np.random.uniform(0, 4*np.pi, n)
y_true = 4 * np.sin(2*x + 0.2)  # Fungsi asli sinus
noise = np.random.normal(0, 0.4, n)
y_noisy = y_true + noise

# Interpolasi Polinomial
# Misal kita pilih orde 5
degree = 5
coeffs = np.polyfit(x, y_noisy, degree)   # cari koefisien polinom
p = np.poly1d(coeffs)                     # bentuk polinom

# Cetak Fungsi Polinom
print("Fungsi polinom hasil interpolasi orde", degree, ":")
print(p)

# Plot Hasil
x_new = np.linspace(0, 4*np.pi, 500)
plt.figure(figsize=(8,4))
plt.scatter(x, y_noisy, color="red", alpha=0.4, label="Data noisy")
plt.plot(x_new, p(x_new), "g-", label=f"Interpolasi Polinom Orde {degree}")
plt.plot(np.sort(x), 4*np.sin(2*np.sort(x) + 0.2), "b--", label="Fungsi Asli")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Interpolasi Polinomial untuk menemukan fungsi asal")
plt.show()
```

### Analisis Interpolasi Orde 5
Interpolasi orde 5 menghasilkan polinom yang bentuknya sudah lebih mirip dengan fungsi asli
dibanding orde lebih rendah. Hal ini menunjukkan bahwa semakin tinggi orde polinomial,
semakin baik pendekatannya terhadap data.


### Kesimpulan
Dari hasil percobaan, interpolasi linear, kuadrat, dan kubik tidak dapat
mengikuti pola sinusoidal. Polinom orde 5 memberikan pendekatan terbaik terhadap
fungsi asal \( f(x) = 4 \sin(2x+0.2) \). Dengan demikian, semakin tinggi orde polinom
yang digunakan, semakin baik hasil interpolasinya.
